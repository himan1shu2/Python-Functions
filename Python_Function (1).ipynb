4.What is the purpose of the return statement in a Python function?
  - What does return do?
     - The return statement sends a result back from a function to where the function was called.
     - It ends the function execution and outputs a value.
     - Without return, a function returns None by default.
  - Why use return?
    - To get output from a function and use it later.
    - To pass calculated values back to the caller.
    - To stop the function once a result is ready.

          def add(a, b):
              return a + b  # Returns the sum
          result = add(3, 5)
          print(result)  # Output: 8

           ##without return
          def add(a, b):
              print(a + b)  # Just prints, no return
          result = add(3, 5)
          print(result)  # Output: None

5.What are iterators in Python and how do they differ from iterables?
   -  Iterable -
        - An iterable is any Python object you can loop over (e.g., using a for loop).
Examples: lists, tuples, strings, dictionaries, sets.

      - Characteristics:

          - Has an __iter__() method that returns an iterator.

          - Can be passed to iter() function to get an iterator.
   
                my_list = [1, 2, 3]  # This is an iterable
                for item in my_list:
                    print(item)

- Iterator -
    - An iterator is an object that produces the next value when you ask for it, one at a time. It keeps track of the current position in the sequence.
    - Characteristics:
        - Has a __next__() method that returns the next item.
        - Raises StopIteration when no more items are left.
        - Has an __iter__() method that returns itself.


           my_list = [1, 2, 3]
           it = iter(my_list)  # 'it' is an iterator

          print(next(it))  # 1
          print(next(it))  # 2
          print(next(it))  # 3
          # next(it) now would raise StopIteration          
6.Explain the concept of generators in Python and how they are defined?
  - Generator -
      - A generator is a special kind of iterator that generates values on the fly, one at a time, only when requested.
      - It does not store all values in memory; it produces them lazily, which makes it memory efficient, especially for large data.
  - How are Generators Defined?
      - There are two main ways to create generators in Python -
         - A. Generator Function (using yield)

               def count_up_to(n):
                   count = 1
                   while count <= n:
                         yield count  # produce current count and pause
                         count += 1
                
               #Using the generator
               counter = count_up_to(5)
               print(next(counter))  # 1
               print(next(counter))  # 2
  
           - B. Generator Expression (like list comprehension but with parentheses)
            
                 squares = (x*x for x in range(5))
                 print(next(squares))  # 0
                 print(next(squares))  # 1


  - Why  Use Generators?
     - Memory efficient: No need to store the entire sequence in memory.
     - Lazy evaluation: Values are produced only when needed.
     - Useful for working with large datasets or infinite sequences.

9.Explain the purpose and usage of the map() function in Python.
  - The map() -
    - function in Python is used to apply a function to every item in an iterable (like a list, tuple, etc.) and return a new iterator with the results.
  - Purpose -
      -  To avoid writing loops for applying the same function to many values.
      - Makes code shorter, cleaner, and faster (especially with lambda functions).

             syntax - map(function, iterable)   
             ## function: A function that processes each item.
             ## iterable: A list, tuple, etc.   
  - Returns what?
       - map() returns a map object (an iterator).
       - You need to use list(), tuple(), or loop to see the results.

  - Usage Example
     - 1: Using map() with a function

           def square(x):
               return x * x
     
           numbers = [1, 2, 3, 4]
           result = map(square, numbers)

           print(list(result))  # Output: [1, 4, 9, 16]

      - 2: Using map() with lambda

              numbers = [1, 2, 3, 4]
              result = map(lambda x: x + 10, numbers)

              print(list(result))  # Output: [11, 12, 13, 14]

10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?   
  - Similarity -
     - Take a function as the first argument
     - Work on iterables (like lists)
     - Return results based on how the function is applied     

  - map() ‚Äì Transforms each item
    - Purpose: Apply a function to every item in an iterable.
    - Returns: A new iterator with transformed items.

```
nums = [1, 2, 3]
result = map(lambda x: x * 2, nums)
print(list(result))  # [2, 4, 6]
 ```
  - filter() ‚Äì Filters items
       - Purpose: Keep items that return True from the function.
       - Returns: A new iterator with filtered items.

```
nums = [1, 2, 3, 4]
result = filter(lambda x: x % 2 == 0, nums)
print(list(result))  # [2, 4]
```

   - reduce() ‚Äì Reduces all items to one
        - Purpose: Repeatedly apply a function to items to reduce them to a single value.
        -  Returns: A single final value.
        - `Must import from functools.`

```
from functools import reduce

nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, nums)
print(result)  # 10
```

````python
#1 . Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

my_list = [10, 3, 5, 6, 8]
print(sum_of_even_numbers(my_list))
````

````python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(17))
print(is_prime(20))
````

````python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_2(n):
    i = 0
    while i <= n:
        yield 2 ** i
        i += 1
for val in powers_of_2(4):
    print(val, end=' ')
````

````python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')
for line in read_file_line_by_line('example.txt'):
    print(line)
````

````python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
````

````python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

input_string = "Hello, how are you?"
result = ''.join(filter(is_not_vowel, input_string))

print(result)
````
